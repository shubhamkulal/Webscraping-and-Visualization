```python
#Import Library
import pandas as pd
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
#Proxy to access Html code
HEADERS = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}
html_text = requests.get('https://studyabroad.shiksha.com/mba-in-abroad-dc11508?uaf[]=country&ctry[]=9&rf=filters',headers=HEADERS)
soup = BeautifulSoup(html_text.content, "lxml")
Universities = soup.find_all('div', class_ = "_1822 _0fc7 _7efa")
```

```python
data = []


for University in Universities:
    University_Name = University.find('div', class_ = "c8ff").text
    
    Location = University.find('span', class_ = "_5588").text
    
    Course = University.find('div', class_ = "dcfd").text
    
    exams_accepted = []
    if University.find('a', title="IELTS"):
        exams_accepted.append('IELTS')
    if University.find('a', title="TOEFL"):
        exams_accepted.append('TOEFL')
    if University.find('a', title="GRE"):
        exams_accepted.append('GRE')
        
    Url = University.find("a",class_="_9865 ripple dark").get("href")
    
    intake_session_ = University.find('label', text='Intake Session')
    if intake_session_:
        intake_session = intake_session_.find_next('li').text.strip()
    else:
        intake_session = None
        
    course_duration_ = University.find('label', text='Course Duration')
    if course_duration_:
        course_duration = course_duration_.find_next('div').text.strip()
    else:
        course_duration = None   
        
    tution_fees_ = University.find('label', text='1st year Tution Fees').find_next('div')
    tution_fees = tution_fees_.text.strip()
    
    work_experience_ = University.find('label',text='Work Experience')
    if work_experience_:
        work_experience = work_experience_.find_next('div').text.strip()
    else:
        work_experience = None
    
    grad_marks_ = University.find('label',text = 'Graduation Marks')
    if grad_marks_:
        grad_marks = grad_marks_.find_next('div').text.strip()
    else:
        grad_marks = None
        
    num_of_schol_ = University.find('label',text = 'No. of Scholarship').find_next('div')
    No_of_Scholarship = num_of_schol_.text.strip()
    
    Shiksha_Grade_ = University.find('span', class_ = "bf6b _1d12")
    if Shiksha_Grade_:
        Shiksha_Grade = Shiksha_Grade_.text
    else:
        Shiksha_Grade = None
        

    print(f"University Name = {University_Name}")
    print(f"Location = {Location}")
    print(f"Courses = {Course}")
    print(f"exams accepted = {exams_accepted}")
    print(f"Url = https://studyabroad.shiksha.com{Url}") 
    print(f"Intake Session = {intake_session}")
    print(f"Course Duration = {course_duration}")
    print(f"First Year Tution Fees = {tution_fees}")
    print(f"Work Experience = {work_experience}")
    print(f"Graduation Marks Required = {grad_marks}")
    print(f"No. of Scholarship Available = {No_of_Scholarship}")
    print(f"Grade = {Shiksha_Grade}")
    print("")
    
    data.append([University_Name, Location, Course, exams_accepted, Url, intake_session, course_duration, tution_fees, work_experience, grad_marks, No_of_Scholarship, Shiksha_Grade])

    df = pd.DataFrame(data, columns=['University_Name', 'Location', 'Course', 'exams_accepted', 'Url', 'intake_session', 'course_duration', 'tution_fees', 'work_experience', 'grad_marks', 'No_of_Scholarship', 'Shiksha_Grade'])

    df.to_excel("shiksha.xlsx")
```
